## ¿Qué tanto tendrían que cobrar los cubanos para llegar a fin de mes?

Imagine por un momento que su salario de un mes entero cabe en el pasillo de una bodega. Que cada mañana, al despertar, no piensa en sus metas o sueños, sino en una ecuación matemática despiadada: ¿Cuántos días de mi sueldo vale un kilo de pollo? ¿A cuántos huevos debo renunciar para poder comprar jabón?

Esta no es una hipótesis abstracta. Es la realidad diaria de millones de cubanos. Mientras la economía global habla de inflación en porcentajes, en la realidad las cosas son distintas. El salario promedio oficial ronda los 6685 pesos cubanos, pero una canasta básica realista supera, sin esfuerzo, los 30,000. La brecha no es un desajuste; es un abismo.

Pero, ¿cómo se cuantifica una crisis que todos sienten pero pocos pueden medir con exactitud? ¿Es posible traducir el "no me alcanza" a un lenguaje irrefutable de datos, gráficos y correlaciones? Este proyecto de ciencia de datos nace precisamente de esa necesidad: dejar de sentir la economía para empezar a entenderla, número por número, producto por producto.

Aquí no usaremos solo estadísticas oficiales. Rastrearemos el precio real en las calles (en las mypimes y en las tiendas estatales en usd) y lo enfrentaremos, sin atenuantes, al poder adquisitivo del salario medio.

Este análisis va más allá de un ejercicio académico. Es un mapa de la resistencia cotidiana. Al visualizar esta brecha, no solo evidenciamos una crisis, sino que aportamos una herramienta precisa para comprender su magnitud, discutir sus causas y, eventualmente, pensar en soluciones.

Acompáñenos a descifrar la ecuación imposible que define la vida en Cuba hoy: la que hay entre el precio en la etiqueta y el salario en el bolsillo. La respuesta, en datos, es más elocuente que cualquier discurso.
