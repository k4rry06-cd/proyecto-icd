## El Dilema del Consumidor: Poder Adquisitivo vs Realidad de Precios en la Economía Cubana Actual

El análisis de la relación entre los salarios nominales y los precios es fundamental para comprender la dinámica del poder adquisitivo de los trabajadores. La evolución de estas variables está condicionada por múltiples factores económicos y estructurales e incluye el crecimiento económico, la productividad laboral y las decisiones de política salarial. 
El índice de precios al consumo (IPC) es un indicador que mide la variación en los precios de bienes y servicios a lo largo del tiempo, reflejando los cambios  en  el  costo  de  vida  de  la población.  Cuando  el  IPC  aumenta,  el  poder adquisitivo de los salarios se reduce, a menos que los ajustes salariales compensen el alza en los precios. La relación entre estas dos variables es crucial, ya que permite determinar si los salarios son suficientes para mantener el nivel de bienestar de los trabajadores o si han experimentado pérdidas debido a la inflación.
En Cuba, desde el 2020,  la inflación ha ido escalando. Las cifras oficiales aportadas por la Oficina Nacional de Estadística e Información (ONEI) indican que en los últimos años el IPC alcanzó niveles críticos. A pesar de la reciente disminución, los expertos advierten que estos números continúan siendo perjudiciales para la economía, especialmente para las familias, trabajadores y pensionistas de menos ingresos y refleja una realidad económica compleja y preocupante para la población. 
En la economía cubana, con una dolarización parcial, el ajuste salarial constituye una vía fundamental de compensación frente a la inflación, lo que resalta la importancia de analizar si los salarios son coherentes con el costo de vida de los cubanos. La relación salario–precio es crucial para determinar si los salarios son suficientes para mantener el bienestar de los trabajadores. 
Este proyecto analiza con datos reales esa realidad cotidiana. Se muestra cuánto del salario se consume en productos básicos y cómo varían los precios entre distintos mercados. Porque entender los números detrás de la economía familiar es el primer paso para tomar mejores decisiones en tiempos de inflación.
